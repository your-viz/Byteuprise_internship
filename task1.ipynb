# Heart Disease Data Visualization
- Age distribution
- Cholesterol distribution
- Gender distribution for heart disease
- Outcome composition

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the Diabetes dataset
df = pd.read_csv('datasets/heart_data.csv')
```

```python
df.head(5) #first 5 rows of dataset
```

```python
np.size(df)
```

```python
heart_disease=df[df['target']==1]
```

```python
# Plotting the Age Distribution
plt.figure(figsize=(8, 5))
plt.hist(heart_disease['age'], bins=20, color='green',edgecolor='black',alpha=0.6)
plt.title('Age Distribution for heart disease')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the cholesterol Distribution

plt.figure(figsize=(8, 5))
plt.hist(heart_disease['chol'], bins=20, color='blue',edgecolor='black',alpha=0.6)
plt.title('Cholesterol Distribution for heart Disease')
plt.xlabel('cholesterol')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the gender distribution for heart disease (categorical data)
outcome_counts = heart_disease['sex'].value_counts()
plt.figure(figsize=(8, 5))
outcome_counts.plot(kind='bar', color=['blue', 'orange'],alpha=0.5)
plt.title('Gender distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['male', 'female'], rotation=0)
plt.show()
```

```python
# Plotting the Outcome Composition (categorical data)
outcome_counts = df['target'].value_counts()
plt.figure(figsize=(8, 5))
outcome_counts.plot(kind='bar', color=['green', 'red'],alpha=0.5)
plt.title('Heart disease Outcome Composition')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No heart disease', 'heart disease'], rotation=0)
plt.show()
```
